```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data\goody.csv')
df
```

```python
fig, ax = plt.subplots()
ax.bar(df['Thing'], df['Cost'])
ax.legend(title='Republicans Are Putting a Band-Aid on an Amputation')
ax.set_ylabel('Amount in Billions (USD)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)
data = np.random.random((50, 50, 50))

fig, ax = plt.subplots()

for i, img in enumerate(data):
    ax.clear()
    ax.imshow(img)
    ax.set_title(f"frame {i}")
    # Note that using time.sleep does *not* work here!
    plt.pause(0.1)
```
